```python
df1['Date']=df1.index
```

```python
df1=df1[['close ']]
df1
```

```python
train = df1.iloc[:2780, :]

train = train.dropna()
train
```

```python
test = df1.iloc[2780:, :]

test
```

Now we are going to create an ARIMA model and will train it with the closing price of the stock on the train data. So let us split the data into training and test set and visualize it.


```python
plt.plot(train.index, train['close '], color='green', label='train')
plt.plot(test.index, test['close '], color='red', label='test')
plt.title('Train vs Test')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

order = (1, 0, 0)  # (p- represents the number of lagged observations ,
                           # d- The differencing order,
                                 # q The moving average order)

model = ARIMA(train, order=order)
model_fit = model.fit()
forecast = model_fit.forecast(steps=len(test))

forecast
```

```python
forecast_values=forecast.values
```

```python
forecast_dates=test.index
forecast_dates
```

```python
forecast_df = pd.DataFrame({'Date': forecast_dates, 'Forecast': forecast_values})

# Set the 'Date' column as the index
forecast_df.set_index('Date', inplace=True)

# Print the combined DataFrame
print(forecast_df)
```

```python
import matplotlib.pyplot as plt

# Plotting training data, testing data, and forecasted values
plt.plot(train, color="red", label="Training Data")
plt.plot(test, color="green", label="Testing Data")
plt.plot(forecast_df['Forecast'], color="blue", label="Forecast")

# Add legend and show the plot
plt.legend()
plt.show()
```
